### Modifing and cleaning all the values

Now we are cleaning each of the single features: converting them to proper scales (0/1) and droping additional values of no interest (Dont know/Not Sure/Refused/Missing).

```python
# 1. var (_MICHD)
# Respondents that have ever reported having coronary heart disease (CHD) or myocardial infarction (MI)
# Transforming our traget value of ever having had a heart attack or disease
# Changing 2 to 0 because this means someone did not have MI (myocardial infarction) or CHD (coronary heart disease)

df_selected['_MICHD'] = df_selected['_MICHD'].replace({2: 0})
df_selected._MICHD.unique()
```

```python
# 2. var (_AGEG5YR) Fourteen-level age category
# Age is already ordinal; 1 equals ages ranging from 18-24 all the way up to 13 which equals to age 80 and older
# The scale increments by 5 years 
# Remove 14 because it is don't know or missing values

df_selected = df_selected[df_selected._AGEG5YR != 14]
df_selected._AGEG5YR.unique()
```

```python
# 3. var (SEX) Indicate sex of respondent
# Changing the scale of sex to 0,1 --> 2 to 0; Female as 0, Male is 1

df_selected['SEX'] = df_selected['SEX'].replace({2:0})
df_selected.SEX.unique()
```

```python
# 4. var (EDUCA)
# What is the highest grade or year of school you completed?
# EDUCA is already an ordinal variable on a scale from 1-6 (1 being never attended school or kindergarten; 
# 6 being college for 4 years or more (graduate)
# Remove 9 for missing/refused values

df_selected = df_selected[df_selected.EDUCA != 9]
df_selected.EDUCA.unique()
```

```python
# 5. var (CHILDREN)
# How many children less than 18 years of age live in your household? Scale 1-87, 88 = None
# Remove 99 refused values / transforming 88 to 0

df_selected = df_selected[df_selected.CHILDREN != 99]
df_selected['CHILDREN'] = df_selected['CHILDREN'].replace({88:0})
df_selected.CHILDREN.unique()
```

```python
# 6. var (_TOTINDA)
# Adults who reported doing physical activity or exercise during the past 30 days other than their regular job
# Changing the scale to 0,1 --> 2 to 0; No exercise as 0, Exercise as 1
# Remove 9 refused values

df_selected = df_selected[df_selected._TOTINDA != 9]
df_selected['_TOTINDA'] = df_selected['_TOTINDA'].replace({2:0})
df_selected._TOTINDA.unique()
```

```python
# 7. var (VETERAN3) Have you ever served on active duty in the United States Armed Forces, 
                  # either in the regular military or in a National Guard or military reserve unit? 
# Changing the scale of Veteran to 0,1 --> 2 to 0; Non-Veteran as 0, Veteran as 1

df_selected = df_selected[df_selected.VETERAN3 != 7]
df_selected = df_selected[df_selected.VETERAN3 != 9]
df_selected['VETERAN3'] = df_selected['VETERAN3'].replace({2:0})
df_selected.VETERAN3.unique()
```

```python
# 8. var (EMPLOY1) Employment status
# EMPLOY1 is already an ordinal variable on a scale from 1-8 (1 being employed for wages; 8 being unable to work)
# Remove 9 for refused values

df_selected = df_selected[df_selected.EMPLOY1 != 9]
df_selected.EMPLOY1.unique()
```

```python
# 9. var (_STATE) State FIPS Code
# no change needed
```

```python
# 10. var (INCOME2) Is your annual household income from all sources: 
# INCOME2 is already ordinal variable on a scale from 1-8 (1 being less than $10,000; 8 being $75,000 or more)
# Remove 77 and 99 for don't know and refused values 

df_selected = df_selected[df_selected.INCOME2 != 77]
df_selected = df_selected[df_selected.INCOME2 != 99]
df_selected.INCOME2.unique()
```

```python
# 11. var (CHECKUP1)
# About how long has it been since you last visited a doctor for a routine checkup?
# CHECKUP1 is already ordinal variable on a scale from 1-8 (1 being within a yesr; 8 being never)
# Remove 9 for refused values 

df_selected = df_selected[df_selected.CHECKUP1 != 9]
df_selected.CHECKUP1.unique()
```

```python
# 12. var (RENTHOM1) --> OWNHOME
# RENTHOM1 is already an ordinal variable 
# Changing the scale of RENTHOM1 to 0,1 --> 2 to 0; Rent home as 0, Own home is 1
# Remove 3, 7, 9 for other arrangements, don't know or refused values

df_selected = df_selected[df_selected.RENTHOM1 != 3]
df_selected = df_selected[df_selected.RENTHOM1 != 7]
df_selected = df_selected[df_selected.RENTHOM1 != 9]
df_selected['RENTHOM1'] = df_selected['RENTHOM1'].replace({2:0})
df_selected.RENTHOM1.unique()
```

```python
# 13. var (MSCODE) Metropolian Status Code
# no changes needed
```

```python
# 14. var (_PRACE1) Preferred race category
# _PRACE1 is already an ordinal variable 
# Remove 7, 8, 77, 99 for No preferred race, Multiracial but preferred race not answered, don't know or refused values

df_selected = df_selected[df_selected._PRACE1 != 7]
df_selected = df_selected[df_selected._PRACE1 != 8]
df_selected = df_selected[df_selected._PRACE1 != 77]
df_selected = df_selected[df_selected._PRACE1 != 99]
df_selected._PRACE1.unique()
```

```python
# 15. var (MARITAL) marital status
# MARITAL is already an ordinal variable 
# Remove 9 for refused values

df_selected = df_selected[df_selected.MARITAL != 9]
df_selected.MARITAL.unique()
```

```python
# 16. var (SMOKE100) Smoked 100 cigarettes in life?
# SMOKE100 is already an ordinal variable 
# Changing the scale of SMOKE100 to 0,1 --> 2 to 0; No as 0, Yes is 1
# Remove 7, 9 for don't know or refused values

df_selected = df_selected[df_selected.SMOKE100 != 7]
df_selected = df_selected[df_selected.SMOKE100 != 9]
df_selected['SMOKE100'] = df_selected['SMOKE100'].replace({2:0})
df_selected.SMOKE100.unique()
```

```python
# 17. var (_BMI5) Body Mass Index (BMI)
# These BMI are multiplied by 100. So e.g. a BMI of 4018 is really 40.18

df_selected['_BMI5'] = df_selected['_BMI5'].div(100).round(0)
df_selected._BMI5.unique()
```

```python
# 18. var (_RFBMI5) obease/overweight ?
# Adults who have a body mass index greater than 25.00 (Overweight or Obese)
# _RFBMI5 is already an ordinal variable 
# Changing the scale of _RFBMI5 to 0,1 --> 2 to 0; No as 0, Yes is 1

df_selected['_RFBMI5'] = df_selected['_RFBMI5'].replace({1:0})
df_selected['_RFBMI5'] = df_selected['_RFBMI5'].replace({2:1})
df_selected._RFBMI5.unique()
```

```python
# 19. var (_RFHYPE5)
# Adults who have been told they have high blood pressure by a doctor
# Changing the scale of _RFHYPE5 to 0,1; No as 0, Yes is 1

df_selected['_RFHYPE5'] = df_selected['_RFHYPE5'].replace({1:0})
df_selected['_RFHYPE5'] = df_selected['_RFHYPE5'].replace({2:1})
df_selected._RFBMI5.unique()
```

```python
# 20. var (TOLDHI2)
# Have you EVER been told by a doctor, nurse or other health professional that your blood cholesterol is high?
# Changing the scale of TOLDHI2 to 0,1; No as 0, Yes is 1
# Remove 7, 9 for don't know or refused values

df_selected = df_selected[df_selected.TOLDHI2 != 7]
df_selected = df_selected[df_selected.TOLDHI2 != 9]
df_selected['TOLDHI2'] = df_selected['TOLDHI2'].replace({2:0})
df_selected.TOLDHI2.unique()
```

```python
# 21. var (_CHOLCHK)
# Cholesterol check within past five years
# _CHOLCHK is already ordinal variable on a scale from 1-2 (1 being yes; 2 being no)
# Changing the scale of _CHOLCHK to 0,1; No as 0, Yes is 1
# Remove 9 for don't know or refused values

df_selected = df_selected[df_selected._CHOLCHK != 9]
df_selected['_CHOLCHK'] = df_selected['_CHOLCHK'].replace({2:0})
df_selected._CHOLCHK.unique()
```

```python
# 22. var (CVDSTRK3)
# (Ever told) you had a stroke
# Changing the scale of CVDSTRK3 to 0,1; No as 0, Yes is 1
# Remove 7, 9 for don't know or refused values

df_selected = df_selected[df_selected.CVDSTRK3 != 7]
df_selected = df_selected[df_selected.CVDSTRK3 != 9]
df_selected['CVDSTRK3'] = df_selected['CVDSTRK3'].replace({2:0})
df_selected.CVDSTRK3.unique()
```

```python
# 23. var (DIABETE3)
# (Ever told) you have diabetes
# Changing the scale of DIABETE3 to 0,1; No as 0, Yes is 1 (counting pre-diabetes/borderline diabetes as diabetes)
# Remove 2(during pregnancy), 7, 9 for don't know or refused values

df_selected = df_selected[df_selected.DIABETE3 != 2]
df_selected = df_selected[df_selected.DIABETE3 != 7]
df_selected = df_selected[df_selected.DIABETE3 != 9]
df_selected['DIABETE3'] = df_selected['DIABETE3'].replace({3:0})
df_selected['DIABETE3'] = df_selected['DIABETE3'].replace({4:1})
df_selected.DIABETE3.unique()
```

```python
# 24. var (_FRTLT1)
# Consume Fruit 1 or more times per day
# Changing the scale of _FRTLT1 to 0,1; No as 0, Yes is 1 
# Remove 9 for don't know or refused values

df_selected = df_selected[df_selected._FRTLT1 != 9]
df_selected['_FRTLT1'] = df_selected['_FRTLT1'].replace({2:0})
df_selected.DIABETE3.unique()
```

```python
# 25. var (_VEGLT1)
# Consume Vegetables 1 or more times per day
# Changing the scale of _VEGLT1 to 0,1; No as 0, Yes is 1 
# Remove 9 for don't know or refused values

df_selected = df_selected[df_selected._VEGLT1 != 9]
df_selected['_VEGLT1'] = df_selected['_VEGLT1'].replace({2:0})
df_selected._VEGLT1.unique()
```

```python
# 26. var (_RFDRHV5)
# Heavy drinkers (adult men having more than 14 drinks per week)
# Changing the scale of _RFDRHV5 to 0,1; No as 0, Yes is 1 
# Remove 9 for don't know or refused values

df_selected['_RFDRHV5'] = df_selected['_RFDRHV5'].replace({1:0})
df_selected['_RFDRHV5'] = df_selected['_RFDRHV5'].replace({2:1})
df_selected = df_selected[df_selected._RFDRHV5 != 9]
df_selected._RFBMI5.unique()
```

```python
# 27. var (HLTHPLN1)
# Do you have any kind of health care coverage ?
# Changing the scale of HLTHPLN1 to 0,1; No as 0, Yes is 1 
# Remove 7, 9 for don't know or refused values

df_selected = df_selected[df_selected.HLTHPLN1 != 7]
df_selected = df_selected[df_selected.HLTHPLN1 != 9]
df_selected['HLTHPLN1'] = df_selected['HLTHPLN1'].replace({2:0})
df_selected.HLTHPLN1.unique()
```

```python
# 28. var (MEDCOST) 
# Was there a time in the past 12 months when you needed to see a doctor but could not because of cost?
# Changing the scale of MEDCOST to 0,1; No as 0, Yes is 1 
# Remove 7, 9 for don't know or refused values

df_selected = df_selected[df_selected.MEDCOST != 7]
df_selected = df_selected[df_selected.MEDCOST != 9]
df_selected['MEDCOST'] = df_selected['MEDCOST'].replace({2:0})
df_selected.MEDCOST.unique()
```

```python
# 29. var (GENHLTH) 
# Would you say that in general your health is: 1-5 scale, 1 being the best
# Remove 7, 9 for don't know or refused values

df_selected = df_selected[df_selected.GENHLTH != 7]
df_selected = df_selected[df_selected.GENHLTH != 9]
df_selected.GENHLTH.unique()
```

```python
# 30. var (MENTHLTH) 
# Mental health, including stress & depression, out of 30 days for how many days was your mental health not good?
# change 88 to 0 for NONE
# Remove 77, 99 for don't know or refused values

df_selected = df_selected[df_selected.MENTHLTH != 77]
df_selected = df_selected[df_selected.MENTHLTH != 99]
df_selected['MENTHLTH'] = df_selected['MENTHLTH'].replace({88:0})
df_selected.MENTHLTH.unique()
```

```python
# 31. var (PHYSHLTH)
# physical health, including physical illness & injury, out of 30 days for how many was your physical health not good?
# change 88 to 0 for NONE
# Remove 77, 99 for don't know or refused values

df_selected = df_selected[df_selected.PHYSHLTH != 77]
df_selected = df_selected[df_selected.PHYSHLTH != 99]
df_selected['PHYSHLTH'] = df_selected['PHYSHLTH'].replace({88:0})
df_selected.PHYSHLTH.unique()
```

```python
# 32. var (DIFFWALK)
# Do you have serious difficulty walking or climbing stairs?
# Changing the scale of DIFFWALK to 0,1; No as 0, Yes is 1 
# Remove 7, 9 for don't know or refused values

df_selected = df_selected[df_selected.DIFFWALK != 7]
df_selected = df_selected[df_selected.DIFFWALK != 9]
df_selected['DIFFWALK'] = df_selected['DIFFWALK'].replace({2:0})
df_selected.DIFFWALK.unique()
```

```python
# 33. var (CHCCOPD1)
# (Ever told) you have Chronic Obstructive Pulmonary Disease or COPD, emphysema or chronic bronchitis?
# Changing the scale of CHCCOPD1 to 0,1; No as 0, Yes is 1 
# Remove 7, 9 for don't know or refused values

df_selected = df_selected[df_selected.CHCCOPD1 != 7]
df_selected = df_selected[df_selected.CHCCOPD1 != 9]
df_selected['CHCCOPD1'] = df_selected['CHCCOPD1'].replace({2:0})
df_selected.CHCCOPD1.unique()
```

```python
# 34. var (MAXVO2_)
# Estimated Age-Gender Specific Maximum Oxygen Consumption
# These MAXVO2 are multiplied by 100. So e.g. a MAXVO2 of 2469 is really 24.69
# Remove 99900 for don't know or refused values

df_selected['MAXVO2_'] = df_selected['MAXVO2_'].div(100).round(0)
df_selected = df_selected[df_selected.MAXVO2_ != 99900]
df_selected.MAXVO2_.unique()
```

```python
# 35. var (_PACAT1)
# Physical Activity Categories
# _PACAT1 is already ordinal variable on a scale from 1-4 (1 being highly active; 8 being inactive)
# Remove 9 for refused values 

df_selected = df_selected[df_selected._PACAT1 != 9]
df_selected._PACAT1.unique()
```

```python
# 36. var (_PAINDX1)
# Physical Activity Index (Meet Aerobic Recommendations ?)
# Changing the scale of _PAINDX1 to 0,1; No as 0, Yes is 1
# Remove 9 for refused values 

df_selected = df_selected[df_selected._PAINDX1 != 9]
df_selected['_PAINDX1'] = df_selected['_PAINDX1'].replace({2:0})
df_selected._PAINDX1.unique()
```

```python
# double chack for NaN
df_selected.isnull().sum(axis = 0)
```

```python
df_final = df_selected.rename(columns = {'_MICHD':'HeartDiseaseOrAttack', '_AGEG5YR':'Age', 'SEX':'Sex', 
                                         'EDUCA':'Education', 'CHILDREN':'Children','_TOTINDA':'PhysActivity',
                                         'VETERAN3':'Veteran','EMPLOY1':'EmploymentStatus', '_STATE':'State', 
                                         'INCOME2':'Income', 'CHECKUP1':'RoutineCeckup', 'RENTHOM1':'OwnHome', 
                                         'MSCODE':'MSCode','_PRACE1':'Race', 'MARITAL':'MaritalStatus', 
                                         'SMOKE100':'Smoker','_BMI5':'BMI', '_RFBMI5':'OverweightOrObese', 
                                         '_RFHYPE5':'HighBP','TOLDHI2':'HighChol', '_CHOLCHK':'CholCheck',
                                         'CVDSTRK3':'Stroke', 'DIABETE3':'Diabetes', '_FRTLT1':'Fruits', 
                                         '_VEGLT1':'Veggies', '_RFDRHV5':'HvyAlcoholConsump', 
                                         'HLTHPLN1':'AnyHealthcare', 'MEDCOST':'NoDocBcCost', 'GENHLTH':'GenHlth', 
                                         'MENTHLTH':'MentHlth', 'PHYSHLTH':'PhysHlth', 'DIFFWALK':'DiffWalk', 
                                         'CHCCOPD1':'COPD', 'MAXVO2_':'MaxVO2', '_PACAT1':'PhysActive', 
                                         '_PAINDX1':'AerobicRec'})
                                         
```
